```python
# Import core libraries

import pandas as pd

# Load the dataset
transactions = pd.read_csv("transactions.csv")
customers = pd.read_csv("customers.csv")
accounts = pd.read_csv("accounts.csv")
complaints = pd.read_csv("complaints.csv")

print("Dataset successfully loaded")
```

#### Customer dataset

```python
customers.info()
```

```python
# Check the dataset for missing values
missing_values_customers = customers.isna().sum()
print(f"Total missing values\n{missing_values_customers}")
```

```python
# Check the dataset for duplicates
duplicate_values_customers = customers.duplicated().sum()
print(f"Total number of duplicated values: {duplicate_values_customers}")
```

```python
# Check the customer_id is in complaint.csv
customer_ids = customers['customer_id']
check_complaints = complaints[complaints['customer_id'].isin(customer_ids)]


print(f"Total Customer_id found in complaint csv: {check_complaints['customer_id'].sum()}")
```

```python
# Confirm transaction_id is in complaint.csv
transaction_ids = transactions['transaction_id']
check_complaints_txn = complaints[complaints['linked_transaction_id'].isin(transaction_ids)]

print(f"Total Customer_id found in complaint csv:\n{check_complaints_txn['linked_transaction_id'].head()}")
```
